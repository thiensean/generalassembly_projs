```python
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
dict_pos_map = {
    # Look for NN in the POS tag because all nouns begin with NN
    'NN': NOUN,
    # Look for VB in the POS tag because all nouns begin with VB
    'VB':VERB,
    # Look for JJ in the POS tag because all nouns begin with JJ
    'JJ' : ADJ,
    # Look for RB in the POS tag because all nouns begin with RB
    'RB':ADV  
}
```

```python
rws = all_reviews_df[['rating', 'rating_description']]
```

```python
rws.head(5)
```

```python
rws['no_contract'] = rws['rating_description'].apply(lambda x: [contractions.fix(token).strip().lower() for token in x.split() if not contractions.fix(token).strip().lower() in stop_words]).str.join(' ')
```

```python
rws['token'] = rws['no_contract'].apply(lambda x: [tokenizer.tokenize(x.strip().lower())])
```

```python
rws['pos'] = rws['token'].apply(lambda x: [pos_tag(token) for token in x])
```

```python
rws.head(5)
```

```python
rws['pos1'] = rws['token'].apply(lambda x: [get_wordnet_pos(pos_tag(token)) for token in x])
```

```python
rws['lemmatized'] = rws['pos'].apply(lambda x: [x[0][0]])
```

```python
rws.head(5)
```

```python
rws['wordnet_pos'] = rws['pos_tags'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
rws.head()
```
